## 1. My lane and why

My provisional lane is **Lane 2: Refresh / Content Opportunity Scoring**.

I chose this lane because I want to help an SEO or content team decide which pages should be reviewed or refreshed first. A website can have many pages, and it is not practical to manually inspect everything. A scoring or ranking system can help prioritize pages that are visible in search, declining, outdated, or underperforming.

This lane fits the starter dataset because the data includes useful signals such as impressions, CTR, average position, content age, days since last update, word count, sessions, and trend direction. These signals can support a practical decision: which pages deserve human review first.

## 2. The question: decision, action, cost of a wrong call

**Search question:**  
Which visible website pages should be prioritized for content refresh because they are likely declining or underperforming?

**Unit of analysis:**  
One content page. Each row in the starter dataset represents one anonymized page/content item.

**Output:**  
A ranked list of pages recommended for review or refresh. Each recommendation should include a score and reason codes, such as stale visible page, declining with demand, low CTR visible page, or engagement review candidate.

**Decision improved:**  
The output helps an SEO or content team decide which pages to inspect first when they have limited time.

**Action someone could take:**  
A reviewer could update outdated content, improve the title or meta description, check whether the page matches search intent, expand thin content, improve engagement elements, protect a page that is at risk, or monitor the page before making changes.

**Cost of a wrong recommendation:**  
A false positive means the team may waste time reviewing a page that does not need attention. A false negative means the team may miss a page that is actually declining or underperforming, which could lead to lost traffic, missed leads, or weaker search visibility.

**Why data or ML can help:**  
Data or ML can help because the decision is not based on only one metric. A page may have high impressions but low CTR, strong position but declining trend, or old content that still receives traffic. A model or scoring system can combine multiple observable signals and rank pages more consistently than manual checking alone.

## 3. Quick look at the data

The next code cell loads the starter dataset and prints a few real numbers that support my lane choice.

```python
# Load the starter data in Google Colab by cloning the starter repo first.

import os
import subprocess
import pandas as pd

REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

# Download the repo only if it is not already available in Colab.
if not os.path.exists(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

# Load the starter dataset.
df = pd.read_csv(f"{REPO_DIR}/data/raw/content_refresh_anonymized.csv")

# Create a simple declining label.
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

# Print supporting numbers.
print("Number of pages:", len(df))
print("Declining rate:", round(df["is_declining_label"].mean(), 3))
print("Correlation between search_volume and impressions_90d:", round(df["search_volume"].corr(df["impressions_90d"]), 3))
```
